# FORMATLAMA

```python
a=3
b=4
print("{}+{}'nin toplamı {}'dır.".format(a,b,a+b))
```

```python
print("{1}+{0}'nin toplamı {2}'dır.".format(a,b,a+b))
```

```python
q=3.1234
w=41.3452
print("{:.2f}+{:.2f}'nin toplamı {:.2f}'dir)".format(q,w,q+w))
```
